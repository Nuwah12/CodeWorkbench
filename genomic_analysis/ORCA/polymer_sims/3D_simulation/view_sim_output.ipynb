```python
%matplotlib widget
import h5py
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.interpolate import splprep, splev
```

```python
# Class to get dataset names from h5 file
class H5ls:
    def __init__(self):
        # Store an empty list for dataset names
        self.names = []

    def __call__(self, name, h5obj):
        # only h5py datasets have dtype attribute, so we can search on this
        if hasattr(h5obj,'dtype') and not name in self.names:
            self.names += [name]
```

```python
coords = []
h5ls = H5ls()
with h5py.File('sim_outs/blocks_0-99.h5','r') as f:
    f.visititems(h5ls)
    t = np.array(f.get(h5ls.names[1]))
t.shape
# Plot 3D points
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(t[:,0],t[:,1],t[:,2])

plt.show()
```
